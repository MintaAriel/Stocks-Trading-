```python
import pytesseract
import PIL.Image
import cv2
import pandas as pd
```

```python
myconfig = r'--psm 6 --oem 3'

def photo_to_txt(photo):
    text = pytesseract.image_to_string(PIL.Image.open(photo), config=myconfig)
    print(text)
    #a = PIL.Image.open(photo)
    return text

def txt_to_df(data_str):
    #Remove the blank lines of the str 
    txt_without_blank = "\n".join(data_str.split("\n\n"))   
    
    # Split the string by newline characters
    parts = txt_without_blank.split('\n')
    
    # Extract the required strings
    part1 = parts[parts.index('<')+1]  # 'AAPL - name of the stock'
    part2 = '\n'.join(parts[parts.index('Operation details')+1:parts.index('Operation details')+7])  # Amount, Shares, Market price, Clearing fee, Total, Date of purchase
    part3 = parts[parts.index('Operation details')+7]  # 'Completed- status'
    part4 = parts[parts.index('<')+2]  #Buy or sell
    
    lines = part2.split('\n')
    
    # Create a dictionary from the lines
    data_dict = {' '.join(line.split(' ')[0:-1]): ' '.join(line.split(' ')[-1:]) for line in lines if line!= lines[5]}  #line 5 Date of purchase
    data_dict[' '.join(lines[5].split(' ')[:3])] = ' '.join(lines[5].split(' ')[3:])
                                                       
    
    data_dict['Stock'] = part1 
    data_dict['Status'] = part3 
    data_dict['B_S'] = part4 
    
    return data_dict 

def dictionaty_to_df(dicti):
    df = pd.DataFrame(dicti, index=[0])
    # Print the DataFrame
    display(df)
    return df
    
text1 = photo_to_txt('photo3.jpg')
dict1 = txt_to_df(text1)
dictionaty_to_df(dict1)
```
