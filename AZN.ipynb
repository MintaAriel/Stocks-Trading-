```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy
import seaborn as sns
```

```python
# Define the ticker symbol
ticker = 'AZN'

# Get the ticker object
stock = yf.Ticker(ticker)

# Get dividends data
dividends = stock.dividends
print("Dividends:")
print(dividends.head())
```

```python
# Convert the Series to a DataFrame
dividends_df = dividends.reset_index()

# Rename the columns
dividends_df.columns = ['Date', 'Dividend']
dividends_df['Year'] = dividends_df['Date'].dt.year
x = np.array([i for i in range(65)])

# Display the DataFrame
print(dividends_df)
dividends_df.to_csv('AZN2024dividends.csv', index=False)
```

```python
plt.plot(dividends_df['Date'], dividends_df['Dividend'], marker = '+' )
```

```python
# Summing dividends by year
dividend_sum_by_year = dividends_df.groupby('Year')['Dividend'].sum().reset_index()

# Displaying the result
print(dividend_sum_by_year)
```

```python
plt.plot(dividend_sum_by_year['Year'], dividend_sum_by_year['Dividend'], marker = '+' )
```
